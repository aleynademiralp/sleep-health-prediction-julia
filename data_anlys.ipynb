```julia
p1 = boxplot(SHLD.Age, 
    title="Yaş", 
    label=false, 
    ylabel="Değer")

p2 = boxplot(SHLD[!, "Sleep Duration"], 
    title="Uyku Süresi", 
    label=false, 
    color=:orange)

p3 = boxplot(SHLD[!, "Heart Rate"], 
    title="Nabız", 
    label=false, 
    color=:green)

p4 = boxplot(SHLD[!, "Daily Steps"], 
    title="Günlük Adım", 
    label=false, 
    color=:purple)


plot(p1, p2, p3, p4, layout=(2, 2), size=(800, 600))
```
