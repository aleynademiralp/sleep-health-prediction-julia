```julia
test_pred=predict(linreg,test)
```

```julia
perf_test = DataFrame(y_original = test[!, "Stress Level"], y_pred = test_pred)
```

```julia
perf_test.hata=perf_test[!,:y_original]-perf_test[!,:y_pred]
```

```julia
perf_test.hata_kare=perf_test.hata.*perf_test.hata
```

```julia
perf_test
```

```julia
train_pred=predict(linreg,train)
```

```julia
perf_train=DataFrame(y_original=train[!, "Stress Level"],y_pred=train_pred)
```

```julia
perf_train.hata=perf_train[!,:y_original]-perf_train[!,:y_pred]
```

```julia
perf_train.hata_kare=perf_train.hata.*perf_train.hata
```

```julia
perf_train
```

```julia
function mape(perf_df)
    mape=mean(abs.(perf_df.hata./perf_df.y_original))
    return mape
end
```

```julia
mape(perf_test)
```

```julia
mape(perf_train)
```

```julia
function rmsd(perf_df)
    rmsd=sqrt(mean(perf_df.hata.*perf_df.hata))
    return rmsd
end
```

```julia
rmsd(perf_test)
```

```julia
rmsd(perf_train)
```

```julia
function mse(perf_df)   
    mse = mean(perf_df.hata .* perf_df.hata)
    return mse
end
```

```julia
mse(perf_test)
```

```julia
mse(perf_train)
```

```julia
#Test hatalarının karşılaştırılması
println("Ortalama Mutlak Test Hatası: ",mean(abs.(perf_test.hata)))            #MAE
println("Ortalama Mutlak Yüzde Test Hatası: ",mape(perf_test))                 #MAPE
println("Kök Ortalama Kare Test Hatası: ", rmsd(perf_test))                    #RMSD
println("Hata Kareleri Ortalaması Test Hatası: ", mean(perf_test.hata_kare))   #MSE
```

```julia
#     MAE:  Kişisin stres seviyesini tahmin ederken +0.6 veya -0.6  puanlık bir yanılma payı ile çalışıyor. 
#     MAPE: Modelin yaptığı hataların gerçek değere oranı. Modelimiz %12 hata payı ile çalışıyor.
#     RMSD:(RMSD-MAE bakılır.) Farkımız çok değil(0.2). Bu durum, modelinin genel olarak istikrarlı olduğunu, ancak bazı tahminlerde ortalamadan biraz
#          daha fazla saptığını gösterir. Yani birkaç veri noktasında model 2-3 puanlık hatalar yapmış olabilir, bu da RMSE'yi biraz yukarı çekmiş.
#     MSE:(=RMSD^2)
```

```julia
histogram(perf_test.hata,title="Test Hata Analizi",ylabel="Frekans",xlabel="Hata",legend=true)
```

```julia
#Train Hataları Karşılaştırması
println("Ortalama Mutlak Train Hatası: ",mean(abs.(perf_train.hata)))            #MAE
println("Ortalama Mutlak Yüzde Train Hatası: ",mape(perf_train))                 #MAPE
println("Kök Ortalama Kare Train Hatası: ", rmsd(perf_train))                    #RMSD
println("Hata Kareleri Ortalaması Train Hatası: ", mean(perf_train.hata_kare))   #MSE
```

```julia
#     MAE: Modelimiz eğitim setindeki verilerle tahmin yaparken ortalama 0.59 puanlık yanılma ile çalışıyor.
#     MAPE: Modelin eğitim verisindeki ortalama hata oranı %11.2. Yani modelimiz verilerin %89'unu doğru açıklayabiliyor.
#     RMSD:(RMSD-MAE bakılır.) Farkın düşük olması (0.18) modelin eğitim setinde büyük ve tutarsız hatalar yapmadığını, istikrarlı olduğunu gösterir.
#     MSE: (RMSD^2)
```

```julia
histogram(perf_train.hata,title="Train Hata Analizi",ylabel="Frekans",xlabel="Hata",legend=false)
```

```julia
#  GENEL SONUÇ: train_rmsd değeri çok düşük ve test_rmsd çok yüksek olsaydı model ezberlemiş (overfitting) olurdu.
#               Bizim bulduğumuz sonuçları kıyasladığımızda modelimizde overfitting yoktur diyebilirz. 
```

```julia
using GLM
fm3 = @formula(var"Stress Level" ~ var"Quality of Sleep" + Age + var"Heart Rate")
mulreg = lm(fm3, train)
```

```julia
#  Prob değerlerini incelediğimzde uyku kalitesi değişkeni ve  kalp atışı hızı değişkenimiz anlamlı yaş değişkenimiz anlamsız çıkmıştır.
#  Diğer değişkenler sabitken yaşın stres seviyesi üzerine ayırt edici etkisi yoktur.
```

```julia
train_kopya = copy(train)
train_kopya[!, "BMI Category"] = strip.(string.(train_kopya[!, "BMI Category"]))
train_kopya[!, "BMI Category"] = replace(train_kopya[!, "BMI Category"], 
    "Normal Weight" => "1.0", 
    "Normal" => "1.0", 
    "Overweight" => "2.0", 
    "Obese" => "3.0")

train_kopya[!, "BMI Category"] = parse.(Float64, train_kopya[!, "BMI Category"])

using GLM
fm4 = @formula(var"Stress Level" ~ var"Quality of Sleep" + var"BMI Category" + var"Heart Rate")
mulreg2 = lm(fm4, train_kopya)
```

```julia
r2(mulreg2)
```

```julia
test_kopya = copy(test)
test_kopya[!, "BMI Category"] = strip.(string.(test_kopya[!, "BMI Category"]))
test_kopya[!, "BMI Category"] = replace(test_kopya[!, "BMI Category"], 
    "Normal Weight" => "1.0", 
    "Normal" => "1.0", 
    "Overweight" => "2.0", 
    "Obese" => "3.0")

test_kopya[!, "BMI Category"] = parse.(Float64, test_kopya[!, "BMI Category"])

test_pred2=predict(mulreg2,test_kopya)
train_pred2=predict(mulreg2,train_kopya)
```

```julia
perf_test2=DataFrame(y_original=test_kopya[!, "Stress Level"],y_pred=test_pred2)
```

```julia
perf_test2.hata=perf_test2[!,:y_original]-perf_test2[!,:y_pred]
```

```julia
perf_test2.hata_kare=perf_test2.hata.*perf_test2.hata
```
